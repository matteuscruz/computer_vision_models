```python
# importando a bilioteca do openCV
import cv2

# Definicao do local do arquivo para conversao
vidcap = cv2.VideoCapture('/content/drive/MyDrive/Faster RCNN (Mãe)/IMG_2975.MOV')
success,image = vidcap.read()

# Comercar o contador como zero e sucesso como verdadeiro (True) para o LOOP
count = 0
success = True

while success:
  # Nome do arquivo que sera gerado
  cv2.imwrite("/content/drive/MyDrive/Faster RCNN (Mãe)/cida%d.jpg" % count, image)     
  success,image = vidcap.read()

  # Leitura do proximo quadro
  print('Read a new frame: ', success)
  
  # Contador para renomear os arquivos
  count += 1
```
